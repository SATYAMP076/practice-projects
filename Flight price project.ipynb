```python
sc=StandardScaler()
x1=sc.fit_transform(x)
```

```python
test_n=sc.fit_transform(test2)
```

```python
from sklearn.decomposition import PCA
for i in range(2,30):
    pca = PCA(n_components=i)
    x_n = pca.fit_transform(x1)
    print(i,sum(pca.explained_variance_ratio_))
```

```python
from sklearn.decomposition import PCA
for i in range(2,29):
    pca = PCA(n_components=i)
    x_test = pca.fit_transform(test_n)
    print(i,sum(pca.explained_variance_ratio_))
```

```python
pca=PCA(n_components=24)
x_n=pca.fit_transform(x1)
```

```python
pca=PCA(n_components=24)
x_test=pca.fit_transform(test_n)
```

```python
from sklearn.linear_model import LinearRegression,SGDRegressor,Lasso,Ridge,BayesianRidge,ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor,XGBRFRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor,BaggingRegressor
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
lr=LinearRegression()
```

```python
for i in range(1,500):
    x_train,x_test,y_train,y_test=train_test_split(x_n,y,test_size=0.30,random_state=i)
    lr.fit(x_train,y_train)
    pred_tr=lr.predict(x_train)
    pred_te=lr.predict(x_test)
    if round(r2_score(y_train,pred_tr)*100,1)==round(r2_score(y_test,pred_te)*100,1):
        print('Random state',i)
        print('Train score',r2_score(y_train,pred_tr)*100)
        print('Test score',r2_score(y_test,pred_te)*100)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_n,y,test_size=0.30,random_state=427)
lr.fit(x_train,y_train)
pred_lr=lr.predict(x_test)
print('train score',lr.score(x_train,y_train)*100)
print('R2 score',r2_score(y_test,pred_lr)*100)
print('RMSE',np.sqrt(mean_squared_error(y_test,pred_lr)))
print('Absolute error',mean_absolute_error(y_test,pred_lr))
```

```python
parameters={'kernel':['poly','rbf','linear'],'C':[0.1,0.01,1,10,0.001]}
clf = GridSearchCV(SVR(), parameters, cv=5,scoring="r2")
clf.fit(x_n,y)
clf.best_params_
```

```python
neighbors={"n_neighbors":range(1,30)}
clf = GridSearchCV(KNeighborsRegressor(),neighbors, cv=5,scoring="r2")
clf.fit(x_n,y)
clf.best_params_
```

```python
pr={'alpha':[0.1,0.01,0.001],'fit_intercept':[True,False],'normalize':[True,False],'precompute':['auto',True,False]}
clf = GridSearchCV(Lasso(), pr,scoring="r2",cv=5)
clf.fit(x_n,y)
print(clf.best_params_)
```

```python
pr={'alpha':[0.1,0.01,0.001],'solver':['auto','svd','cholesky','lsqr','sparse_cg','sag','saga','lbfgs']}
clf = GridSearchCV(Ridge(), pr,scoring="r2",cv=5)
clf.fit(x_n,y)
print(clf.best_params_)
```

```python
dtc=DecisionTreeRegressor()
svr=SVR(kernel="linear",C=10)
kn=KNeighborsRegressor(n_neighbors=7)
l1=Lasso(alpha=0.01,fit_intercept=True,normalize=True,precompute=False)
l2=Ridge(alpha=0.01,solver="sag")
l3=ElasticNet()
l4=BayesianRidge()
sgd=SGDRegressor()
xgb=XGBRegressor()
xgbr=XGBRFRegressor()
```

### Regularization

```python
def regul(f):
    f.fit(x_train,y_train)
    print(f,'\n\t',f.score(x_train,y_train)*100)
    pred=f.predict(x_test)
    print('R2_score :',r2_score(y_test,pred)*100)
    print('error1:\n:',mean_absolute_error(y_test,pred))
    print('RSME:\n:',np.sqrt(mean_squared_error(y_test,pred)))
```

```python
regul(l1)
```

```python
regul(l2)
```

```python
regul(l3)
```

```python
regul(l4)
```

### apply more regressor

```python
def regressor(f):
    f.fit(x_train,y_train)
    print(f,'\n',f.score(x_train,y_train)*100)
    pred=f.predict(x_test)
    print('error1:\n',mean_absolute_error(y_test,pred))
    print('RSME:\n',np.sqrt(mean_squared_error(y_test,pred)))
    print('r2 score:\n',r2_score(y_test,pred)*100)
```

### apply ensemble methods

```python
ensemble=[RandomForestRegressor(),AdaBoostRegressor(),GradientBoostingRegressor(),BaggingRegressor()]
```

```python
for i in ensemble:
    i.fit(x_train,y_train)
    print('\n \nscore: ',i,':\n',i.score(x_train,y_train)*100)
    pred=i.predict(x_test)
    print(' R2 score:',r2_score(y_test,pred)*100)
    print('error1:\n',mean_absolute_error(y_test,pred))
    print('RSME:\n',np.sqrt(mean_squared_error(y_test,pred)))
```

#### hypertuning the best models

```python
pr={'n_neighbors':range(1,10),'weights':['uniform','distance'],'algorithm':['auto','ball_tree','kd_tree','brute'],'metric':['str','callable','minkowski'],'p':[1,2]}
grid = GridSearchCV(KNeighborsRegressor(),pr,cv=5,scoring='r2')
grid.fit(x_n,y)
print(grid.best_params_)
print(grid.best_score_)
```

```python
knn=KNeighborsRegressor(algorithm='auto',n_neighbors=5,weights='uniform',metric='minkowski',p=1)
knn.fit(x_train,y_train)
pred_knn=knn.predict(x_test)
print("score",knn.score(x_train,y_train)*100)
print('R2_score :',r2_score(y_test,pred_knn)*100)
print('error1:\n:',mean_absolute_error(y_test,pred_knn))
print('RSME:\n:',np.sqrt(mean_squared_error(y_test,pred_knn)))
```

```python
for i in range(2,15):
    crs_score=cross_val_score(knn,x_n,y,cv=i,scoring='r2')
    score=crs_score.mean()
    print('CV value',i)
    print('CV score',score*100)
    print('r2 score',r2_score(y_test,pred_knn)*100)
```

```python
scores=cross_val_score(knn,x_n,y,cv=11)
score=np.mean(scores)
std=np.std(scores)
print('CV mean',score)
print('std:',std)
```
