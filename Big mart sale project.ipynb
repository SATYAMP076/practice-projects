### filling null values

```python
def fill_size(row):
    if(pd.isnull(row['Outlet_Size'])):
        if(row['Outlet_Type']=="Supermarket Type1"):
            row['Outlet_Size']="Small"
            return row
        elif(row['Outlet_Type']=="Supermarket Type2"):
            row['Outlet_Size']="Medium"
            return row
        elif(row['Outlet_Type']=="Supermarket Type3"):
            row['Outlet_Size']="Medium"
            return row
        else:
            row['Outlet_Size']="Small"
            return row
    else:
        return row
```

```python
train=train.apply(fill_size,axis=1)
test=test.apply(fill_size,axis=1)
```

```python
def fill_null(row):
    if(pd.isnull(row['Item_Weight'])):
        row['Item_Weight']=train.groupby('Item_Type').mean()['Item_Weight'][row['Item_Type']]
        return row
    else:
        return row
```

```python
train=train.apply(fill_null,axis=1)
test=test.apply(fill_null,axis=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split,RandomizedSearchCV,GridSearchCV
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
lr=LinearRegression()
```

#### select best random state

```python
for i in range(1,500):
    x_train,x_test,y_train,y_test=train_test_split(x1,y,test_size=0.30,random_state=i)
    lr.fit(x_train,y_train)
    pred_tr=lr.predict(x_train)
    pred_te=lr.predict(x_test)
    if round(r2_score(y_train,pred_tr)*100,1)==round(r2_score(y_test,pred_te)*100,1):
        print('Random state',i)
        print('Train score',r2_score(y_train,pred_tr)*100)
        print('Test score',r2_score(y_test,pred_te)*100)
```

```python
lr=LinearRegression()
x_train,x_test,y_train,y_test=train_test_split(x1,y,test_size=0.30,random_state=424)
lr.fit(x_train,y_train)
print('Intercept:',lr.intercept_)
print('\nTrain score:',lr.score(x_train,y_train)*100)
pred_lr=lr.predict(x_test)
print('R2_score: ',r2_score(y_test,pred_lr)*100)
print('\nError1: ',mean_absolute_error(y_test,pred_lr))
print('Error2 RMSE: ',np.sqrt(mean_squared_error(y_test,pred_lr)))
```
