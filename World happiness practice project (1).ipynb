# Feature engineering

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
x['Region']=lb.fit_transform(x['Region'])
x
```

# Remove skewness

```python
from sklearn.preprocessing import power_transform,StandardScaler
```

```python
x=power_transform(x,method='yeo-johnson')
```

# Use scaler method

```python
sc=StandardScaler()
x=sc.fit_transform(x)
```

# Model making

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
lr=LinearRegression(fit_intercept=True,normalize=False,positive=False)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=20)
lr.fit(x_train,y_train)
print('Intercept:',lr.intercept_)
print('Coef: ',lr.coef_)
print('\nTrain score:',lr.score(x_train,y_train)*100)
pred=lr.predict(x_test)
print('R2_score: ',r2_score(y_test,pred)*100)
print('\nError1: ',mean_absolute_error(y_test,pred))
print('Error2 RMSE: ',np.sqrt(mean_squared_error(y_test,pred)))
```

```python
for i in range(0,1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=i)
    lr.fit(x_train,y_train)
    pred_tr=lr.predict(x_train)
    pred_te=lr.predict(x_test)
    if round(r2_score(y_train,pred_tr)*100,1)==round(r2_score(y_test,pred_te)*100,1):
        print('\nRandom_state',i,'perform well')
        print('R2_score',r2_score(y_train,pred_tr)*100)
        print('R2_score',r2_score(y_test,pred_te)*100)

    
```

we taking random_state=68 with getting highest score in both train test sets

```python
from sklearn.model_selection import GridSearchCV
param1={'fit_intercept':[True,False],'normalize':[True,False],'positive':[True,False]}
m1=LinearRegression()
grid=GridSearchCV(estimator=m1,param_grid=param1)
grid.fit(x_train,y_train)
print('Score: ',grid.best_score_)
print('Best parmatr: ',grid.best_params_)
```

# Ensemble method

```python
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,GradientBoostingRegressor
from sklearn.model_selection import cross_val_score
```

```python
ensemble=[RandomForestRegressor(),AdaBoostRegressor(),GradientBoostingRegressor()]
```

```python
for i in ensemble:
    i.fit(x_train,y_train)
    print('\n score: ',i,':\n',i.score(x_train,y_train))
    pred=i.predict(x_test)
    print(' R2 score:',r2_score(y_test,pred))
    print('error1:\n',mean_absolute_error(y_test,pred))
    print('RSME:\n',np.sqrt(mean_squared_error(y_test,pred)))
    
```
